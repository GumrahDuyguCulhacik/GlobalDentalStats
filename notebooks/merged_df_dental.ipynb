| Sütun Adı                        | Açıklama                                                                 |
|----------------------------------|--------------------------------------------------------------------------|
| IndicatorCode                    | Göstergenin kodu (örneğin: ORALHEALTH_DENTALHEALTHCARE_EXP_TOTAL)       |
| Indicator                        | Göstergenin açıklaması (örneğin: Diş sağlığı harcamaları toplamı)       |
| ValueType                        | Değerin türü (örneğin: numeric)                                          |
| ParentLocationCode               | Bölge kodu (örneğin: AFR = Afrika)                                       |

| Location type                    | Coğrafi birimin türü (örneğin: Country)                                  |
| SpatialDimValueCode              | Ülke kodu (örneğin: AGO = Angola)                                        |
| Location                         | Ülke adı (örneğin: Angola)                                               |
| Period type                      | Zaman türü (örneğin: Year)                                               |
| Period                           | Veri yılı (örneğin: 2019)                                                |
| FactValueUoM                     | Ölçü birimi (örneğin: million US$)                                       |
| FactValueNumericLowPrefix        | Alt sınır ön eki                                         |
| FactValueNumericLow              | Sayısal alt sınır değeri                                                 |
| FactValueNumericHighPrefix       | Üst sınır ön eki                                         |
| FactValueNumericHigh             | Sayısal üst sınır değeri                                                 |
| Value                            | Göstergenin gerçek değeri (örneğin: 70.2 milyon USD)                     |
| FactValueTranslationID           | Çeviri ID si (genellikle boş)                                            |
| FactComments                     | Ek açıklamalar (genellikle boş)                                          |
| Language                         | Veri dili (örneğin: EN)                                                  |
| DateModified                     | Verinin son güncellenme tarihi (örneğin: 2023.04.20)                     |

```python
df12["label"] =  df12.groupby("Location").cumcount().map({0: "both_cancer", 1: "male_cancer", 2: "female_cancer"})

df_wide= df12.pivot(index= "Location", columns= "label", values= "Value").reset_index()

df_wide.columns.name= None

df_wide.head()
```

```python
#  Gerekli sütunları seçip yeniden adlandır
df1_renamed = df1[['Location', 'Period', 'Value']].rename(columns={'Value': 'TotalExpenditure'})
df3_renamed = df3[['Location', 'Period', 'Value']].rename(columns={'Value': 'PerCapitaExpenditure'})
df4_renamed = df4[['Location', 'Period', 'Value']].rename(columns={'Value': 'EdentulismRate'})
df5_renamed = df5[['Location', 'Period', 'Value']].rename(columns={'Value': 'PeriodontalRate'})
df6_renamed = df6[['Location', 'Period', 'Value']].rename(columns={'Value': 'UntreatedCariesRate'})
df13_renamed = df13[['Location', 'Period', 'Value']].rename(columns={'Value': 'UntreatedDeciduousCaries11_9Rate'})
df14_renamed = df14[['Location', 'Period', 'Value']].rename(columns={'Value': 'ProductivityLossRate'})


# 3️Merge işlemleri
merged_df = df1_renamed.merge(df3_renamed, on=['Location', 'Period'])
merged_df = merged_df.merge(df4_renamed, on=['Location', 'Period'])
merged_df = merged_df.merge(df5_renamed, on=['Location', 'Period'])
merged_df = merged_df.merge(df6_renamed, on=['Location', 'Period'])
merged_df = merged_df.merge(df13_renamed, on=['Location', 'Period'])
merged_df = merged_df.merge(df14_renamed, on=['Location', 'Period'])

merged_df = merged_df.merge(df_wide, on='Location', how='left')

```
